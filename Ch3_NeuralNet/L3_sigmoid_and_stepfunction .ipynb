## Sigmoid function & Step function 

```python
import numpy as np 
import matplotlib.pyplot as plt 


def step_function(x):
    y = x > 0
    return y.astype(np.int)


def sigmoid(x):
    y = 1 / (1 + np.exp(-x))
    return y
```

```python
x         =  np.arange(-5.0, 5.0, 0.1)
step_y    =  step_function(x)
sigmoid_y =  sigmoid(x)
```

```python
# Plotting 
plt.plot(x, step_y)
plt.plot(x, sigmoid_y)
plt.ylim(-0.1, 1.1)    #y축의 범위 
plt.show()
```
