# 편미분

```python
%matplotlib notebook

import numpy as np 
import matplotlib.pylab as plt 
from mpl_toolkits import mplot3d    # 3D plotting
```

```python
def numerical_diff(f, x):
    h = 1e-4               # 0.0001 (조낸 작은 수 예시)
    return (f(x+h) - f(x-h)) / (2*h)
```

## Multivariable function 

<center>$f(x_{0}, x_{1})\quad =\quad x_{0}^{2}\quad +\quad x_{1}^{2}$ </center> <br/>

```python
def function_2(x0, x1):
    a    =  x0**2  +  x1**2
    return a 
```

## 정의역(domain)

```python
x0    =    np.arange(-4, 4, 0.1)
x1    =    np.arange(-4, 4, 0.1)
```

```python
X, Y    = np.meshgrid(x0,x1)
```

## 치역(range, =image) 

```python
Z     =    function_2(X, Y)
```

## 3차원 그래프

```python
fig  =  plt.figure()
ax   = plt.axes( projection = '3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')

ax.set_xlabel("x0")
ax.set_ylabel("x1")
ax.set_zlabel("f(x_0, x_1)")
```

# 위의 2변수 함수 그래프를 편미분

주의할 점은 변수가 2개라는 사실.  <br/>
따라서,'어느 변수에 대한 미분이냐'를 구별해야함 


(1) $x_{0}$ 에 대한 미분이냐? <br/>

(2) $x_{1}$ 에 대한 미분이냐? <br/>

이와 같이 변수가 여럿인 함수에 대한 미분을 <span style="color:blue">**편미분**</span>이라고 함

(p. 126) <br/>
$x_{0}=3$, $x_{1}=4$ 일 때, <br/>

$x_{0}$에 대한 편미분 구하기  

```python
def function_tmp1(x0):
    return x0**2 + 4**2             # f(x0, 4) = x0**2 + 4**2 
```

```python
x0 = 3.0
numerical_diff(function_tmp1, x0)   # 수치 미분 
```

$x_{0}=3$, $x_{1}=4$ 일 때, <br/>

$x_{1}$에 대한 편미분 구하기 

```python
def function_tmp2(x1):
    return 3**2 + x1**2             # f(3, x1) = 3**2 + x1**2 
```

```python
x1 = 4.0
numerical_diff(function_tmp2, x1)
```

이처럼 편미분은 변수가 하나인 미분과 마찬가지로 <br/>
**(목표 변수만 살리고 나머지 변수는 고정시킨) 특정 장소의 기울기를 구함**
