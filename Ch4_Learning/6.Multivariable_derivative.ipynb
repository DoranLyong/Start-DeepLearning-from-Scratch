```python
%matplotlib notebook

import numpy as np 
import matplotlib.pylab as plt 
from mpl_toolkits import mplot3d    # 3D plotting
```

## Multivariable function 

<center>$f(x_{0}, x_{1})\quad =\quad x_{0}^{2}\quad +\quad x_{1}^{2}$ </center> <br/>

```python
def function_2(x0, x1):
    a    =  x0 ** 2
    b    =  x1 ** 2
    return a + b 
```

## 정의역(domain)

```python
x0    =    np.arange(-4, 4, 0.1)
x1    =    np.arange(-4, 4, 0.1)
```

```python
X, Y    = np.meshgrid(x0,x1)
```

## 치역(range, =target set) 

```python
Z     =    function_2(X, Y)
```

## 3차원 그래프

```python
fig  =  plt.figure()
ax   = plt.axes( projection = '3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
```
